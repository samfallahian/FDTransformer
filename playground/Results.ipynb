```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
df = pd.read_csv(f"../logs/2023-03-11 155605-cae.csv", index_col=False)
```

```python
df
```

```python
fig,ax = plt.subplots(1,figsize=(12,6))
e = 10
# ax.plot(df["epoch"],df["total_loss"],'o-',markerfacecolor='w',markersize=2)
ax.plot(df[df["epoch"]>e]["epoch"],df[df["epoch"]>e]["total_loss"],'o-',markerfacecolor='w',markersize=2)
ax.plot(df[df["epoch"]>e]["epoch"],df[df["epoch"]>e]["mse"],'o-',markerfacecolor='w',markersize=2)
ax.legend(["Total Loss","MSE"])
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
plt.show()
```
