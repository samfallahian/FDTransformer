```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
```

```python
class Autoencoder(nn.Module):
    def __init__(self, input_dim, encoder_dims, decoder_dims, bilinear_output_dim):
        super(Autoencoder, self).__init__()

        # Encoder
        encoder_layers = []
        prev_dim = input_dim
        for dim in encoder_dims:
            encoder_layers.append(nn.Linear(prev_dim, dim))
            encoder_layers.append(nn.ReLU())
            prev_dim = dim
        self.encoder = nn.Sequential(*encoder_layers)

        # Bi-Linear Layer
        self.bilinear = nn.Bilinear(encoder_dims[-1], encoder_dims[-1], bilinear_output_dim)

        # Decoder
        decoder_layers = []
        prev_dim = bilinear_output_dim
        for dim in decoder_dims:
            decoder_layers.append(nn.Linear(prev_dim, dim))
            decoder_layers.append(nn.ReLU())
            prev_dim = dim
        decoder_layers.append(nn.Linear(prev_dim, input_dim))
        decoder_layers.append(nn.Sigmoid())
        self.decoder = nn.Sequential(*decoder_layers)

    def forward(self, features, labels):
        encoded_features = self.encoder(features)
        encoded_labels = self.encoder(labels)
        bi_linear_output = self.bilinear(encoded_features, encoded_labels)
        decoded_output = self.decoder(bi_linear_output)
        return decoded_output
```

```python
# Example usage
input_dim = 784
encoder_dims = [256, 128, 64]
decoder_dims = [64, 128, 256]
bilinear_output_dim = 64

model = Autoencoder(input_dim, encoder_dims, decoder_dims, bilinear_output_dim)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training loop
for epoch in range(10):
    for batch_features, _ in train_loader:
        optimizer.zero_grad()
        batch_features = batch_features.view(batch_features.size(0), -1)
        batch_labels = torch.randn_like(batch_features)
        print(batch_labels)
        output = model(batch_features, batch_labels)
        loss = criterion(output, batch_features)
        loss.backward()
        optimizer.step()
    print(f"Epoch: {epoch+1}, Loss: {loss.item():.4f}")
```
